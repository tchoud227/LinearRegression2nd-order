```python
# Initialize vectors here
m = len(x)
X = np.zeros((m,3))
Y = np.zeros((m,1))
theta = np.zeros((3,1))
```

```python
# matrices X and Y
# x degree 2, X degree 1, x degree 0
X = np.column_stack([x**2, x, np.ones((m,1))])
# Y values
Y = y

# closed form vector solution
# M = [XTX]^-1(XT)
M = np.dot(np.linalg.inv(np.dot(X.T,X)),X.T)
# theta = MY or  = [XTX]^-1(XTY)
theta = np.dot(M,Y)
print('theta:\n' ,theta)

# y^ = thetaX
yHat = np.dot(X,theta)

# min cost
yMin = np.min(yHat)
print('\nminimum cost:\n',yMin)

#plot data and line fit
plt.plot(x,yHat,'g')
plt.plot(x,y,'o b')
plt.show()
```
